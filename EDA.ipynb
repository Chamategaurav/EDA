Univariate Analysis

## Histogram
It is used to see the bucket-wise distribution of a continuous variable

```python
sns.histplot(df["temp"],bins=[0,5,10,15,20,25,30,35,40])
```

```python
for i in cont_cols:
    print(i)
    sns.histplot(df[i])
    plt.show()
```

### Countplot
It is used to see the count-wise distribution of a categorical variable

```python
sns.countplot(df["workingday"])
```

```python
df["workingday"].value_counts()
```

```python
for i in cat_cols:
    print(i)
    sns.countplot(df[i])
    plt.show()
```

### Boxplot
- It is used to see the quartile-wise distribution of a continuous variable
- It is also used to see the outliers present in the data
- It is used to see the quartile-wise distribution of a continuous variable against a categorical variable

```python
sns.boxplot(df["temp"])
```

```python
df["temp"].describe()
```

```python
for i in cont_cols:
    print(i)
    sns.boxplot(df[i])
    plt.show()
```

#### upper_whisker --> 75_per + 1.5*(75_per - 25_per)
#### lower_whisker --> 25_per - 1.5*(75_per - 25_per)
#### (75_per - 25_per)--> IQR (Inter-Quartile-Range)

### Bivariate Analysis

```python
sns.boxplot(df["mnth"],df["temp"])
```

```python
ls=[]
for i in cat_cols:
    for j in cont_cols:
        print(i,"Vs",j)
        ls.append((i,j))
        sns.boxplot(df[i],df[j])
        plt.show()
ls
```

### ScatterPlot
It is used to see the relationship between two continuous varible

```python
sns.scatterplot(df["temp"],df["cnt"])
```

```python
for i in cont_cols:
    for j in cont_cols:
        if i!=j:
            print(i,"Vs",j)
            sns.scatterplot(df[i],df[j])
            plt.show()
```

```python
sns.pairplot(df[cont_cols])
```

### Barplot
- It is used to see statistical distribution (mean/median/var/std/max/min etc) of a continuous variable against a categorical variable

```python
sns.barplot(df["mnth"],df["temp"],ci=None)
## by default it shows mean
```

```python
df[df["mnth"]==4]["temp"].max()-df[df["mnth"]==4]["temp"].min()
```

```python
df[df["mnth"]==8]["temp"].max()-df[df["mnth"]==8]["temp"].min()
```

```python
df.groupby(["mnth"])["temp"].mean()
```

```python
sns.barplot(df["mnth"],df["temp"],ci=None,estimator=np.median)
```

```python
sns.barplot(df["mnth"],df["temp"],ci=None,estimator=np.max)
```

```python
sns.barplot(df["mnth"],df["temp"],ci=None,estimator=np.min)
```

```python
ls=[]
for i in cat_cols:
    for j in cont_cols:
        print(i,"Vs",j)
        ls.append((i,j))
        sns.barplot(df[i],df[j],ci=None)
        plt.show()
ls
```
